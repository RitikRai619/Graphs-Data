## Rating avg by count


```python
data["Day"] = data["Timestamp"].dt.date

d1=day_average=data.groupby(["Day"]).mean()
plt.figure(figsize=(26,6))
plt.plot(d1.index,d1["Rating"])
```

## Rating Avg by week

```python
data["Week"] = data["Timestamp"].dt.strftime("%Y-%U")
d2=week_average=data.groupby(["Week"]).mean()
plt.figure(figsize=(26,6))
plt.plot(d2.index,d2["Rating"])
```

##  Avg Rating by Month

```python
data["Month"] = data["Timestamp"].dt.strftime("%Y-%m")
d3=month_average=data.groupby(["Month"]).mean()
plt.figure(figsize=(26,6))
plt.plot(d3.index,d3["Rating"])
```

### Avg Rating my Month By Course


```python
data["Month"] = data["Timestamp"].dt.strftime("%Y-%m")
d4=month_average_crs=data.groupby(["Month","Course Name"])["Rating"].mean().unstack()
d4.plot(figsize=(30,8))

```

### What day are people most happy

```python
data["Weekday"] = data["Timestamp"].dt.strftime("%A")
data["Daynumber"] = data["Timestamp"].dt.strftime("%w")
weekday_average=data.groupby(["Weekday","Daynumber"]).mean()
weekday_average= weekday_average.sort_values("Daynumber")
weekday_average.index.get_level_values(0)
plt.figure(figsize=(26,6))
plt.plot(weekday_average.index.get_level_values(0),weekday_average["Rating"])
```
